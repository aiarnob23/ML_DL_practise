```python
salary_dummies = pd.get_dummies(df.salary,dtype=int)
salary_merged = pd.concat([df,salary_dummies],axis='columns')
```

```python
dept_dummies = pd.get_dummies(df.Department,dtype=int) 
final = pd.concat([salary_merged,dept_dummies],axis='columns')
```

```python
salary_merged = pd.concat([df,salary_dummies],axis='columns')
final=pd.concat([df,dept_dummies],axis='columns')
```

```python
final = final.drop(columns=['salary','Department'])
```

```python
final
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = final[final.columns.difference(['left'])]
y = final['left']  

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.2)

model = LogisticRegression(max_iter=1000)

model.fit(x_train, y_train)
```
